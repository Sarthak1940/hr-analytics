```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("/content/hr-analytics.csv")
df.head()
```

```python
df["Gender"] = df["Gender"].astype("category")
df["Attrition"] = df["Attrition"].astype("category")
df["OverTime"] = df["OverTime"].astype("category")
```

```python
df["AttritionFlag"] = (df["Attrition"] == "Yes").astype(int)
df["OverTimeFlag"] = (df["OverTime"] == "Yes").astype(int)
```

```python
df['AgeBand']      = pd.cut(df['Age'], bins=[17,25,30,40,50,60], labels=["18–25","26–30","31–40","41–50","51–60"])
df['TenureBand']   = pd.cut(df['YearsAtCompany'], bins=[-1,1,3,5,10,40], labels=["<1","1–3","3–5","5–10","10+"])
df['IncomeBand']   = pd.qcut(df['MonthlyIncome'], q=5, labels=["Q1","Q2","Q3","Q4","Q5"])
```

```python
satisfaction_cols = ['EnvironmentSatisfaction','JobSatisfaction','RelationshipSatisfaction','WorkLifeBalance']
df['SatisfactionMean'] = df[satisfaction_cols].mean(axis=1)
```

```python
df.head()
```

```python
attr_rate = df["AttritionFlag"].mean()
attr_rate
```

```python
dept_rate = df.groupby("Department")["AttritionFlag"].mean().sort_values(ascending=False)
dept_rate
```

```python
role_rate = df.groupby("JobRole")["AttritionFlag"].mean().sort_values(ascending=False)
role_rate
```

```python
sns.countplot(data=df, x="Attrition")
plt.title("Attrition Distribution")
plt.plot()
```

```python
sns.barplot(data=df, x="Department", y="AttritionFlag")
plt.title("Attrition by Department")
plt.plot()
```

```python
sns.barplot(data=df, x="JobRole", y="AttritionFlag")
plt.title("Attrition by Job Role")
plt.xticks(rotation=90)
plt.show()
```

```python
sns.barplot(data=df, x="OverTime", y="AttritionFlag")
plt.title("Attrition vs Overtime")
plt.show()
```

```python
sns.boxplot(data=df, x="Attrition", y="MonthlyIncome")
plt.title("Attrition vs Income")
plt.show()
```
